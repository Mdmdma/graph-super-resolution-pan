```python
file_path = '/scratch2/merler/code/data/pan10/evaluation_results/results_eval_with_graph.csv'
df = pd.read_csv(file_path)
df = df.drop_duplicates()
df = df.sort_values(by=['batch_size', 'scaling_factor'])
df.reset_index(drop=True, inplace=True)
df.drop('crop_size', axis=1, inplace=True)
df.drop('subset', axis=1, inplace=True)
df.drop('sdi', axis=1, inplace=True)
#df.drop('checkpoint', axis=1, inplace=True)
df.drop('optimization_loss', axis=1, inplace=True)
df.drop('mu', axis=1, inplace=True)
df.drop('average_link', axis=1, inplace=True)
df = df[df['training_mode'] == 'graph-plus']
models = df['training_mode'].unique()
df.rename(columns={'training_mode': 'model'}, inplace=True)
df = df[df['batch_size'] == 4]
pd.set_option('display.max_colwidth', None)
df_gp= df
df_gp
```

```python

# Concatenate the DataFrames vertically
df = pd.concat([ df_b, df_n, df_g, df_gp], axis=0)
df.rename(columns={'bw_error': 'mse_p'}, inplace=True)
df.rename(columns={'downsapled_color_error': 'mse_c'}, inplace=True)
df.rename(columns={'l1_loss': 'mae'}, inplace=True)
df.rename(columns={'mse_loss': 'mse'}, inplace=True)

df
```

```python
import matplotlib.pyplot as plt

# List of metrics
metrics = ['mae', 'mse', 'psnr', 'mse_c', 'mse_p','ergas', 'sam', 'lambda']

plot_parameter = 'model'
plot_parameter_values = df[plot_parameter].unique()
print(plot_parameter_values)
# Iterate over each metric
i = 0

fig, ax = plt.subplots(2, 4, figsize=(20, 10))


for metric in metrics:
    # Create a new figure for each metric
    
    
    # Iterate over each upsampler
    for plot_para in plot_parameter_values:
        # Filter the data for the current upsampler
        filtered_data = df[df[plot_parameter] == plot_para]
        
        # Convert the scaling_factor column to numeric type
        filtered_data['scaling_factor'] = pd.to_numeric(filtered_data['scaling_factor'])
        
        # Plot the metric against scaling_factor
        ax[i//4, i%4].plot(filtered_data['scaling_factor'], filtered_data[metric], label=plot_para)
    
    # Add labels and legend
    ax[i//4, i%4].set_xlabel('Scaling Factor')
    ax[i//4, i%4].set_ylabel(metric)
    ax[i//4, i%4].set_title(f'{metric}')
    ax[i//4, i%4].legend()
    ax[i//4, i%4].legend(title=plot_parameter)
    i+=1
    
    # Show the plot
plt.show()
```
